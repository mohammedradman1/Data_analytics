# Summary of Classifiers

This notebook shows the summary of the various classifiers acting on the datasets.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'classifier_metrics_summary.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(df.head())
```

```python
df_copy = df[~df["Algorithm"].str.contains("KNN")]
# Separate the data based on the 'Technique' column
holdout_df = df_copy[df['Technique'] == 'holdout']
cross_val_df = df_copy[df['Technique'] == 'crossValidation']

# Function to plot metrics for a given technique
def plot_metrics(data, technique_name):
    metrics = data['Metric'].unique()
    algorithms = data['Algorithm'].unique()
    dataset_sizes = data['Dataset_Size'].unique()

    for metric in metrics:
        plt.figure(figsize=(10, 6))
        for algo in algorithms:
            algo_data = data[(data['Metric'] == metric) & (data['Algorithm'] == algo)]
            plt.plot(algo_data['Dataset_Size'], algo_data['Value'], marker='o', label=algo)
        
        plt.title(f'{metric} vs Dataset Size ({technique_name})')
        plt.xlabel('Dataset Size')
        plt.ylabel(metric)
        plt.legend()
        plt.grid(True)
        plt.show()

# Plot the metrics for holdout technique
plot_metrics(holdout_df, 'Holdout')

# Plot the metrics for cross-validation technique
plot_metrics(cross_val_df, 'Cross Validation')
```
