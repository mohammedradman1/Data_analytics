# Evaluation Metrics

```python
# Reading Current Metrics
metrics_df = pd.read_csv("classifier_metrics_summary.csv")
metrics_df.drop(metrics_df.columns[0], axis = 1, inplace = True)

def add_metrics( metric_name, algorithm_name, metrics_values, technique):
  global metrics_df
  # Create an empty list to store the datapoints
  data_list = []
  # Loop through the data dictionary
  for key, value in metrics_values.items():
    # Create a new list for each datapoint with metric and algorithm
    data_point = [ metric_name ,algorithm_name, value, key, technique  ]
    # Append the datapoint to the data_list
    metrics_df.loc[len(metrics_df)] = data_point
```

# Support Vector Machine (SVM) Classifier

## Using holdout

We construct the decision tree classifier and find its accuracy and other measures.

```python
# We import the classifier model and metrics
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
# to store the accuracy, precision
svc_models_accuracy = {}
svc_models_precision = {}
svc_models_recall = {}

for key, value in data_dic.items():
  print("Evaluating ", key , " data.")
  # we split the data
  Y = value["infected"]
  X = value.drop("infected", axis = 1)
  x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)

  # we create model
  svc = SVC(  )

  # we fit the model
  svc.fit(x_train, y_train)

  # predicting the values
  y_pred = svc.predict(x_test)

  # evaluation model accuracy, precision, recall
  svc_models_accuracy[key] = accuracy_score(y_test, y_pred)
  svc_models_precision[key] = precision_score(y_test, y_pred)
  svc_models_recall[key] = recall_score(y_test, y_pred)
```

```python
# print the recall
svc_models_recall
```

We notice that the recall at the small dataset is very high and it dropped significantly as we increase the dataset size. Nomatter how bigger the dataset gets the recall is approximately the same.

```python
add_metrics("Accuracy", "SVM", svc_models_accuracy, "holdout" )
add_metrics("Precision", "SVM", svc_models_precision, "holdout")
add_metrics("Recall", "SVM", svc_models_recall, "holdout")
metrics_df.sample(5)
```

## Cross-validation

```python
# import cross-validation model
from sklearn.model_selection import cross_validate, KFold
from sklearn.metrics import make_scorer # to get accuracy, precision, and recall
```

```python
# specify the scoring we wish to calculate
scoring = {'accuracy' : make_scorer(accuracy_score),
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score)}
```

```python
# to store the accuracy, precision
svc_models_accuracy = {}
svc_models_precision = {}
svc_models_recall = {}

for key, value in data_dic.items():
  print("Evaluating ", key , " data.")
  # we split the data
  target = value["infected"]
  features = value.drop("infected", axis = 1)

  # we create model
  svc = SVC( random_state = 42 )

  # we specify number of folds within the data
  kfold = KFold( n_splits = 10)

  # we fit the model and obtain the results
  model_metrics =  cross_validate(estimator=svc,
                                    X= features,
                                    y=target,
                                    cv=kfold,
                                    scoring=scoring)
  # finding metrics: accuracy, precision, recall
  svc_models_accuracy[key] = model_metrics["test_accuracy"].mean()
  svc_models_precision[key] = model_metrics["test_precision"].mean()
  svc_models_recall[key] = model_metrics["test_recall"].mean()
```

```python
# we print the accuracies
svc_models_accuracy
```

```python
# we print the precision
svc_models_precision
```

```python
# print the recall
svc_models_recall
```

It is noticeable that even with cross-validation, the valuse of accuracy, precision, and recall still decreases with the increase in the dataset size. <br>
And the accuracy and precision become fixed after the 5k size.

```python

add_metrics("Accuracy", "SVM", svc_models_accuracy, "crossValidation" )
add_metrics("Precision", "SVM", svc_models_precision, "crossValidation")
add_metrics("Recall", "SVM", svc_models_recall, "crossValidation")
metrics_df.sample(5)
```
