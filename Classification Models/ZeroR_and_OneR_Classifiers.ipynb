# ZeroR Rule Classifier

We can calculate the majority of the target class and consider it the predicted class for any new value coming.

```python
# We implement the classifier
class ZeroRClassifier():
  def __init__(self):
    self.major_class_value = None # to hold the class of the most common class
    self.accuracy = None

  def fit(self, target_class):
    '''
      Parameter:
        target_class = the target class as a pandas series
        Return: None
    '''
    try:
      self.major_class_value = target_class.mode()[0]
      self.accuracy = len(target_class[ target_class == self.major_class_value ]) /len(target_class)
    except Exception as e:
      print("Error in finding target class: ", e)

  def get_accuracy(self):
    return self.accuracy
```

```python
# to store the accuracy, precision
zeroR_models_metrics = {}

for key, value in data_dic.items():
  print("Evaluating ", key , " data.")
  # we split the data
  Y = value["infected"]

  # we create model
  zeroR_Classifier = ZeroRClassifier()

  # we fit the model
  zeroR_Classifier.fit(Y)

  # store the metrics for this datasize in the dictionary
  zeroR_models_metrics[key] = zeroR_Classifier.get_accuracy()
```

```python
# Print the accuracies
zeroR_models_metrics
```

We notice that the samllest dataset has the highest accuracy, however, for further increase in the size of the dataset, we  barely see a difference in the accuracy. This result may be due to bais in the target class value in the first dataset.

# OneR Classifier

```python
# Import the needed library
from mlxtend.classifier import OneRClassifier
from sklearn.model_selection import train_test_split
```

```python
# to store the accuracy, precision
oneR_models_metrics = {}

for key, value in data_dic.items():
  print("Evaluating ", key , " data.")
  # we split the data
  Y = value["infected"]
  X = value.drop("infected", axis = 1)
  x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
  # we convert the multi-dimension dataframes into numpy arrays
  x_train = x_train.to_numpy()
  x_test = x_test.to_numpy()

  # we create model
  oneR_classifier= OneRClassifier()

  # we fit the model
  oneR_classifier.fit(x_train, y_train)

  # we print the selected feature for the rule and classification rules
  print("For dataset ", key, " feature index ", oneR_classifier.feature_idx_, " has been chosen")
  # Uncomment the below to view the rules
  # print("Rules are\n", oneR_classifier.prediction_dict_, "\n")

  # store the metrics for this datasize in the dictionary
  oneR_models_metrics[key] = oneR_classifier.score(x_test, y_test)
```

```python
# predicting the accuracy of each model
oneR_models_metrics
```

The accuracy is close to the one we obtained from the ZeroR classifier but slightly better for the small dataset.
